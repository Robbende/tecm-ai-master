#### 5.
Implementacion de funcion que calcula la distribucion de frecuencias de las etiquetas en una categoria dentro del corpus `Brown`

```python
def explora_category_by_tagset(category, tagset=None):
    
    palabras_etiq = brown.tagged_words(categories=category, tagset=tagset)
    
    # lista de las etiquetas dentro de las palabras etiquetadas
    tags = [tag for _, tag in palabras_etiq]
    
    # distribucion de frecuencias de las etiquetas
    etiqFD = nltk.FreqDist(tags)
    
    # cantidad de tipos de etiquetas
    num_etiquetas = etiqFD.B()
    
    # lista de las 10 etiquetas mas frecuentes y su frecuencia
    top_tags = etiqFD.most_common(10)
    
    return num_etiquetas, top_tags
```

```python
# prueba de la funcion
explora_category_by_tagset(category='science_fiction', tagset='universal')
```

#### 7.
Entrenamiento y evaluacion del etiquetador HMM

```python
def entrenamientoHMM(oracion, tamanio):
    
    oraciones_etiq = brown.tagged_sents(categories='news')
    
    # datos de entrenamiento
    datos_train = oraciones_etiq[-tamanio:]
    
    # datos de prueba
    datos_test = oraciones_etiq[:100]
    
    # entrenamiento del etiquetador HMM con datos de entrenamiento
    tagger = nltk.HiddenMarkovModelTagger.train(datos_train)
    
    # etiquetado de la oracion
    oracion_etiquetada_hmm = tagger.tag(oracion)

    # evaluacion del accuracy con datos de prueba
    acc = tagger.accuracy(datos_test)
    
    return tagger, oracion_etiquetada_hmm, acc
```

```python
def evaluarHMM():
    oraciones_etiq = brown.tagged_sents(categories='news')
    oracion = [tp[0] for tp in oraciones_etiq[76]]
    
    tagger, oracion_etiquetada_hmm, acc = entrenamientoHMM(oracion, 500)
    print("Entrenando el modelo nltk.HiddenMarkovModelTagger con 500 oraciones...")
    print(f"Oracion etiquetada con el modelo: {oracion_etiquetada_hmm}\n")
    print(f"Accuracy del conjunto de pruebas: {100.0*acc}\n\n\n")
    
    tagger, oracion_etiquetada_hmm, acc = entrenamientoHMM(oracion, 3000)
    print("Entrenando el modelo nltk.HiddenMarkovModelTagger con 3000 oraciones...")
    print(f"Oracion etiquetada con el modelo: {oracion_etiquetada_hmm}\n")
    print(f"Accuracy del conjunto de pruebas: {100.0*acc}\n\n\n")
                                       
```

```python
# evaluacion del modelo
evaluarHMM()
```

### Analisis de los resultados:
El accuracy del modelo HMM aumento considerablemente al entrenar el modelo con una mayor cantidad de oraciones, así mismo el etiquetado de la oración que se pasa a la función cambia ligeramente para algunas palabras haciendo su etiquetado más preciso.
