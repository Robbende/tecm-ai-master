```python
# librerias necesarias
from sklearn.tree import DecisionTreeRegressor
from sklearn import datasets
```

```python
# Conjunto de datos
iris = datasets.load_iris()
caracteristicas = iris.data
objetivos = iris.target
```

```python
# Creacion del arbol de regresion
arbol_regresion = DecisionTreeRegressor(random_state=0)
```

```python
# Entrenamiento del modelo
modelo = arbol_regresion.fit(caracteristicas, objetivos)
```

```python
# Nueva muestra 
muestra = [[ 5, 4, 3, 2]]


# Prediccion de la clase de la nueva muestra
modelo.predict(muestra)
```

```python
# Load libraries
import pydotplus
from IPython.display import Image
from sklearn import tree


# Create DOT data
dot_data = tree.export_graphviz(arbol_regresion, 
                                out_file=None,
                                feature_names=iris.feature_names,
                                class_names=iris.target_names)

# Draw graph
graph = pydotplus.graph_from_dot_data(dot_data)

# Show graph
Image(graph.create_png())
```
