```python
# librerias necesarias
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets

# Conjunto de datos
iris = datasets.load_iris()
caracteristicas = iris.data
objetivos = iris.target


# Creacion del arbol de clasificacion 
arbol_decision = DecisionTreeClassifier(random_state=0)


# Entrenamiento del modelo
modelo = arbol_decision.fit(caracteristicas, objetivos)
```

```python
# Nueva muestra 
muestra = [[ 4, 5, 4, 3]]


# Prediccion de la clase de la nueva muestra
modelo.predict(muestra)
```

```python
# Probabilidades de la clase predicha
modelo.predict_proba(muestra)
```

```python
# Uso del criterio de la entropia
arbol_decision_entropia = DecisionTreeClassifier(criterion='entropy', random_state=0)

# Entrenamiento del modelo
modelo_entropia = arbol_decision_entropia.fit(caracteristicas, objetivos)
```

```python
# Nueva muestra 
muestra = [[ 5, 4, 3, 2]]


# Prediccion de la clase de la nueva muestra
modelo_entropia.predict(muestra)
```

```python
# Load libraries
import pydotplus
from IPython.display import Image
from sklearn import tree


# Create DOT data
dot_data = tree.export_graphviz(arbol_decision, 
                                out_file=None,
                                feature_names=iris.feature_names,
                                class_names=iris.target_names)

# Draw graph
graph = pydotplus.graph_from_dot_data(dot_data)

# Show graph
Image(graph.create_png())
```
