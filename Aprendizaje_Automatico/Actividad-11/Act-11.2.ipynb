```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from sklearn.decomposition import FastICA
from scipy.io import wavfile
```

#### Impotacion del audio

```python
sampling_rate, audio = wavfile.read('wav/Aeropuerto_1.wav')
```

```python
audio.shape
```

#### FastICA

```python
ica = FastICA(n_components=2, whiten='unit-variance')
S_ = ica.fit_transform(audio)
A_ = ica.mixing_
```

```python
S_.shape
```

```python
A_
```

#### Ploteo de las Señales

```python
tiempo = np.linspace(0,8,audio.shape[0])
tiempo
```

```python
# Señal de voz
plt.plot(tiempo,S_[:,0])
plt.show()
```

```python
# Ruido
plt.plot(tiempo,S_[:,1])
plt.show()
```

```python
print(np.allclose(audio, np.dot(S_,A_.T)+ica.mean_))
```
